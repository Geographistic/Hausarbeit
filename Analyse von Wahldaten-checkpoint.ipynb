```r
install.packages("dplyr")
install.packages("ggplot2")
library(dplyr)
library(ggplot2)

```

```r
plot(Strukturdaten$Bevölkerung_mitMigrationshintergrund,Wahldaten$AFD_Prozent)
```

```r
Daten <- as.data.frame(cbind(Strukturdaten$Bevölkerung_mitMigrationshintergrund,Wahldaten$AFD_Prozent))

names(Daten)[1] <- "Bevölkerung_mitMigrationshintergrund"
names(Daten)[2] <- "AFD_Prozent"

Daten

linreg <- lm(Bevölkerung_mitMigrationshintergrund ~ AFD_Prozent, Daten)
    linreg
    summary(linreg)  
```

```r

# this is the predicted line of multiple linear regression
ggplot(data = Daten, aes(x = Bevölkerung_mitMigrationshintergrund, y = AFD_Prozent)) + 
  geom_point(color='blue') +
  geom_line(color='red',data = Daten)

ggplot(data = Daten, aes(x = Bevölkerung_mitMigrationshintergrund, y = AFD_Prozent)) + 
  geom_point(color='blue') +
  geom_smooth(method = "lm", se = FALSE)
```
